```python
# Implementasi Queue menggunakan list
class QueueList:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
      if not self.is_empty():
        return self.queue.pop(0)
      else:
        return None

    def front(self):
      return self.queue[0] if not self.is_empty() else None

    def rear(self):
      return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
      return len(self.queue) == 0

    def size(self):
      return len(self.queue)

# Contoh penggunaan
queue = QueueList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah queue:",queue.queue)
print("Deque element:",queue.dequeue())
print("Front element:",queue.front())
print("Rear element:",queue.rear())
print("Queue size:",queue.size())
```

```python
from collections import deque

# Implementasi Queue mengguanakan deque
class QueueDeque:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        return self.queue.popleft() if not self.is_empty() else None

    def front(self):
        return self.queue[0] if not self.is_empty() else None

    def rear(self):
        return self.queue[-1] if not self.is_empty() else None

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

# Contoh penggunaan
queue = QueueDeque()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Queue setelah queue:",queue.queue)
print("Deque element:",queue.dequeue())
print("Front element:",queue.front())
print("Rear element:",queue.rear())
print("Queue size:",queue.size())
```

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

class QueueLinkedList:
  def __init__(self):
    self.front = self.rear = None

  def enqueue(self, data):
    new_node = Node(data)
    if self.rear is None:
      self.front = self.rear = new_node
      return
    self.rear.next = new_node
    self.rear = new_node

  def dequeue(self):
    if self.front is None:
      return None
    dequeued_element = self.front.data
    self.front = self.front.next
    if self.front is None:
      self.rear = None
    return dequeued_element

  def is_empty(self):
    return self.front is None

  def front_element(self):
    return self.front.data if self.front else None

  def rear_element(self):
    return self.rear.data if self.rear else None

# Contoh penggunaan
queue = QueueLinkedList()
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)
print("Dequeued element:", queue.dequeue())
print("Front element:", queue.front_element())
print("Rear element:", queue.rear_element())
```

```python
class AntrianBioskop:
  def __init__(self, kapasitas = 10):
    self.antrian = []
    self.kapasitas = kapasitas

  def enqueue(self, ticket_id):
    if len(self.antrian) < self.kapasitas:
      if '7' in ticket_id:
          print("Pelanggan VIP, silahkan langsung masuk")
      else:
          self.antrian.append(ticket_id)
          print(f"{ticket_id} masuk kedalam antrian")
    else:
      print("Antrian penuh")

  def dequeue(self):
    if not self.is_empty():
      pelanggan = self.antrian.pop(0)
      print(f"Pelanggan {pelanggan} dilarang masuk")
    else:
      print("Antrian kosong")

  def front(self):
    if not self.is_empty():
      return self.antrian[0]
    else:
      return None

  def rear(self):
    if not self.is_empty():
      return self.antrian[-1]
    else:
      return None

  def is_empty(self):
    return len(self.antrian) == 0

  def size(self):
    return len(self.antrian)

print("=====================================")
print("|      Daftar Antrian Bioskop       |")
print("=====================================")

# Contoh penggunaan
antrian_bioskop = AntrianBioskop()
while True:
  antrian_bioskop.enqueue(input("\nMasukkan nomor antrian: "))
  if input("Apakah ingin melanjutkan?(y/n): ") == "n":
    break

print("\nPelanggan pertama",antrian_bioskop.front())
print("Pelanggan terakhir",antrian_bioskop.rear())
print("\nJumlah pelanggan dalam antrian",antrian_bioskop.size())
```

# **LATIHAN 3**

```python
class AntrianParkir:
    def __init__(self, kapasitas=3):
        self.antrian = []
        self.kapasitas = kapasitas

    def enqueue(self, plat_nomor, kendarann_listrik):
        if len(self.antrian) >= self.kapasitas:
            print("Parkiran penuh, mohon tunggu kendaraan lain keluar")
            return

        if kendarann_listrik.lower() == 'y':
            self.antrian.insert(0, plat_nomor)
            print(f"Kendaraan listrik {plat_nomor} masuk parkiran (Prioritas).")
        else:
            self.antrian.append(plat_nomor)
            print(f"Kendaraan {plat_nomor} masuk parkiran.")

    def dequeue(self):
        if self.is_empty():
            print("Parkiran kosong.")
            return None
        plat_nomor = self.antrian.pop(0)
        print(f"Kendaraan {plat_nomor} keluar parkiran.")
        return plat_nomor

    def front(self):
        if self.is_empty():
            print("Parkiran kosong.")
            return None
        print(f"Kendaraan pertama dalam antrian: {self.antrian[0]}")
        return self.antrian[0]

    def rear(self):
        if self.is_empty():
            print("Parkiran kosong.")
            return None
        print(f"Kendaraan terakhir dalam antrian: {self.antrian[-1]}")
        return self.antrian[-1]

    def is_empty(self):
        return len(self.antrian) == 0

    def size(self):
        print(f"Jumlah kendaraan dalam antrian: {len(self.antrian)}")
        return len(self.antrian)

# Contoh Penggunaan
parkiran = AntrianParkir()

print("======================================")
print("| Selamat Datang di Parkiran Untidar |")
print("======================================")
while True:
    print("\nMenu: ")
    print("1. Tambah kendaraan ke parkiran")
    print("2. Keluarkan kendaraan dari parkiran")
    print("3. Lihat kendaraan pertama dalam antrian")
    print("4. Lihat kendaraan terakhir dalam antrian")
    print("5. Lihat jumlah kendaraan dalam parkiran")
    print("6. Keluar")

    pilihan = input("Pilih menu: ")

    if pilihan == "1":
        plat_nomor = input("Masukkan nomor plat kendaraan: ")
        prioritas = input("Apakah kendaraan listrik? (y/n): ")
        parkiran.enqueue(plat_nomor, prioritas)
    elif pilihan == "2":
        parkiran.dequeue()
    elif pilihan == "3":
        parkiran.front()
    elif pilihan == "4":
        parkiran.rear()
    elif pilihan == "5":
        parkiran.size()
    elif pilihan == "6":
        print("Terima kasih, program selesai.")
        break
    else:
        print("Pilihan tidak valid, silakan coba lagi.")
```

```python
class AntrianSertifikat:
    def __init__(self):
        self.antrian = []
        self.antrian_reguler = []

    def enqueue(self, student_id):
        if '5' in student_id:  # Mahasiswa berprestasi langsung dilayani
            print(f"Nomor antrean {student_id} (Mahasiswa Prestasi) langsung dilayani tanpa mengantri.")

        elif len(self.antrian) < 20:
            self.antrian.append(student_id)
            print(f"Nomor antrean {student_id} ditambahkan ke antrian utama.")
        else:
            self.antrian_reguler.append(student_id)
            print(f"Nomor antrean {student_id} ditambahkan ke antrian reguler.")

    def dequeue(self):
        if self.antrian:
            student_id = self.antrian.pop(0)
            print(f"Nomor antrean {student_id} dilayani (antrian utama).")
            return student_id
        elif self.antrian_reguler:
            student_id = self.antrian_reguler.pop(0)
            print(f"Nomor antrean {student_id} dilayani (antrian reguler).")
            return student_id
        else:
            print("Antrian kosong.")
            return None

    def size(self):
        total_size = len(self.antrian) + len(self.antrian_reguler)
        print(f"\nJumlah mahasiswa dalam antrian: {total_size}")
        return total_size

# Inisialisasi antrian
antrian = AntrianSertifikat()

# Meminta input jumlah mahasiswa
print("=======================================")
print("|   Antrean Pengambilan Sertifikat    |")
print("=======================================")
jumlah_mahasiswa = int(input("Masukkan jumlah antrean: "))
print("")

for i in range(jumlah_mahasiswa):
    npm = str(i + 1).zfill(2)  # Format NPM dengan angka urut (01, 02, dst.)
    student_id = "TI" + npm  # Format ID dengan "TI" di depan
    antrian.enqueue(student_id)

# Menampilkan informasi antrian
antrian.size()

# Melayani mahasiswa pertama dalam antrian
antrian.dequeue()

# Menampilkan informasi antrian setelah dilayani
antrian.size()
```
